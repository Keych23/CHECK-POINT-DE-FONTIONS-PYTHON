```python
# Question 1
def multiplier_liste(liste):
    resultat = 1
    for nombre in liste:
        resultat *= nombre
    return resultat

# Ma Liste d'éléments
Ma_liste = [2, 4, 3]
resultat = multiplier_liste(Ma_liste)

print("Ma liste =", Ma_liste)
print("Résultat =", resultat)
```

```python
# Question 2
def tri(liste):
    """Trie la liste par le dernier élément de chaque tuple"""
    return sorted(liste, key=lambda x: x[-1])

# Liste donnée
liste_tuples = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]

# Utilisation la fonction tri
liste_triee = tri(liste_tuples)

print("Liste originale :", liste_tuples)
print("Liste triée :", liste_triee)
```

```python
# Question 3
# Dictionnaires donnés
d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 300, 'b': 200, 'd': 400}

# Combinaison des dictionnaires
resultat = d1.copy()

for cle, valeur in d2.items():
    if cle in resultat:
        resultat[cle] += valeur
    else:
        resultat[cle] = valeur

# Afficher le résultat
print("Dictionnaire 1:", d1)
print("Dictionnaire 2:", d2)
print("Dictionnaire combiné:", resultat)
```

```python
# Question 4
def generer_dict_carres(n):
    """Génèrons et affichons un dictionnaire des carrés de 1 à n"""
    # Créer le dictionnaire
    carres = {i: i*i for i in range(1, n + 1)}

    # Affichage en formatage amélioré
    print(f"Dictionnaire des carrés de 1 à {n}:")
    print("{", end="")

    # Affichage de chaque paire clé-valeur
    for i, carre in carres.items():
        if i < n:
            print(f"{i}: {carre}, ", end="")
        else:
            print(f"{i}: {carre}", end="")

    print("}")

    return carres

# Entrée de programme : 8
generer_dict_carres(8)
```

```python
# Question 5
# Liste donnée
liste = [('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5')]

# Trier par élément flottant (ordre croissant)
liste_triee = sorted(liste, key=lambda x: float(x[1]))

# Afficher le résultat
print("Liste originale :", liste)
print("Liste triée (croissant) :", liste_triee)

# Pour un tri décroissant
liste_triee_decroissant = sorted(liste, key=lambda x: float(x[1]), reverse=True)
print("Liste triée (décroissant) :", liste_triee_decroissant)
```
